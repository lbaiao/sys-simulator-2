```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import torch

def db_to_power(x):
    return 10**(x/10)
# %matplotlib inline
```

# Testing data

```python
FILE_NAME = 'script43_exec'
N_AGENTS = 2
filepath = f'D:\\Dev\\sys-simulator-2\\data\\dql\\{FILE_NAME}.pickle'
# filepath = 'D:\\Dev\\sys-simulator-2\\data\\a3c\\scratch7.pickle'
file = open(filepath, 'rb')
data = pickle.load(file)
file.close()
```

## MUE success rate

```python
data['mue_success_rate']
```

## MUE Spectral Efficiencies

```python
data['mue_sinr_threshold']
```

```python
mue_speffs = data['mue_speffs']
speff_threshold = np.log2(1 + db_to_power(data['mue_sinr_threshold'])) 
baseline = speff_threshold * np.ones(len(mue_speffs))
plt.plot(mue_speffs)
plt.plot(baseline)
plt.ylabel('Spectral efficiency [bps/Hz]')
plt.xlabel('Timestep')
```

## D2D Summed Spectral Efficiencies

```python
d2d_speffs = data['d2d_speffs']
plt.plot(d2d_speffs)
plt.ylabel('Summed Spectral efficiency [bps/Hz]')
plt.xlabel('Timestep')
```

## Summed rewards

```python
rewards = data['rewards']
plt.plot(rewards)
plt.ylabel('Reward')
plt.xlabel('Timestep')
```

## Chosen actions

```python
actions = data['chosen_actions']
plt.hist(actions)
plt.title('actions')
```
