## Proving why the random search is the worst scheme

### Random search

```python
rnd_search_path = 'D:\Dev\sys-simulator-2\data\\benchmarks\script3\\20210207-115227\log.pickle'
rnd_data = load_with_pickle(rnd_search_path)
```

```python
rewards = rnd_data['rewards']
pathlosses = rnd_data['pathlosses_d2d_to_bs']

fig2, ax1 = plt.subplots(figsize=(7,5))
ax1.set_xlabel('Time step', fontsize='x-large', )
ax1.set_ylabel('Reward', color='tab:blue', fontsize='x-large')
ax1.plot(rewards, '-.', color='tab:blue', markersize=10)
ax1.tick_params(axis='both', labelsize=12)

ax2 = ax1.twinx()
ax2.set_ylabel('Pathloss [dB]', color='tab:red', fontsize='x-large')
ax2.plot(pathlosses, '-.', color='tab:red', markersize=10)
ax2.tick_params(axis='both', labelsize=12)
fig2.tight_layout()
```
